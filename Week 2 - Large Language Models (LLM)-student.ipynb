## What is a Large Language Model (LLM)?

A **large language model (LLM)** is a type of artificial intelligence trained on massive amounts of text (books, articles, websites, code, etc.). Instead of “understanding” language like a human, it learns statistical patterns in how words and sentences are used. By predicting the most likely *next word* in a sequence, an LLM can generate fluent text, answer questions, summarize, translate, write code, and much more. It feels conversational, but under the hood, it’s pattern recognition at scale.

### How It Works (at a high level)

Think of the process like this:

![LLM Visualization](LLMVisualization.png)

Think of the model like a very advanced version of “autocomplete” on your phone. It has read a huge amount of text (like books, articles, websites), so it has learned patterns about how words usually fit together.

When you type something in, the model doesn’t “think” the way people do. Instead, it breaks your text into little pieces called tokens. A token is usually a word, or sometimes part of a word (like “run” and “ning” might be two tokens).

Then the model looks at those tokens and predicts the most likely next token, then the next one, and the next one, until it builds up a complete answer.

## What is Prompt Engineering?

*Prompt engineering* is the process of crafting and refining text prompts to elicit specific, high-quality responses from artificial intelligence models, particularly large language models. It involves designing and optimizing the text input (the prompt) that you provide to an AI model. The goal is to guide the model to generate the most relevant, accurate, and useful output for a given task. It’s like giving clear instructions to a tool to get the desired result.


### Why Does Prompt Engineering Matter?

Large Language Models generate text one piece at a time, measured in tokens (chunks of words or characters). For example, the sentence “ChatGPT is fun!” can be 4 tokens: "ChatGPT", "is", "fun", and "!". Most free and paid models limit how many tokens you can use in a single interaction. Efficient prompts not only save time but also reduce the number of tokens you spend — which matters when models have usage caps or costs.

- **Enhanced Accuracy**: Clear, well-structured prompts help the AI understand the context and requirements, leading to more accurate responses.

- **Efficiency**: Thoughtful prompt design can minimize back-and-forth, saving both time and tokens.

- **Customization**: Prompts can be tailored to specific tasks — from writing and coding to brainstorming or problem-solving — making the AI more useful for your needs.

## Prompting Techniques

To structure prompts for an LLM more effectively, we use _prompt engineering techniques_.

Let's look at some common examples with the following code snippet and its error message:

> **Problem Setup (Code Snippet):**
>
> ```python
>      # Python code with a bug
>      print(total + 5)
> ```
>
> Error:
>
> ```
>      NameError: name 'total' is not defined
> ```


### 1. Zero-Shot Debugging

**Description**: You describe the issue directly, without giving any examples. This relies on the LLM’s prior knowledge.

> **Prompt Example (to LLM):**
>
> I’m getting an error: "NameError: name 'total' is not defined".
> What does this mean and how can I fix it?
>
> **LLM Output (expected):**
>
> The error means that the variable 'total' is used before being defined.
> To fix it, you need to declare and assign a value to 'total' before using it:
> total = 0
> print(total + 5)



### 2. Few-Shot Debugging

**Description**: You give the LLM a few examples of similar bugs and their solutions before asking it to solve a new one. This helps the model understand the pattern.

> **Prompt Example (to LLM):**
>
> Here are some examples of Python errors and their fixes:
>
> Example 1:
> Error: NameError: name 'x' is not defined
> Fix: Define x before using it, e.g., x = 0
>
> Example 2:
> Error: IndexError: list index out of range
> Fix: Check the list length before accessing an index
>
> Now, I have this error:
> "NameError: name 'total' is not defined"
> Can you suggest a fix?
>
> **LLM Output (expected):**
>
> The error occurs because 'total' is used before it is defined.  
> To fix it, define 'total' first, for example:
> total = 0
> print(total + 5)



### 3. Iterative Prompting

**Description**: You interact with the LLM over multiple steps, refining the prompt based on its previous responses to narrow down the solution.

> **Prompt Example (to LLM):**
>
> Step 1: I'm getting a NameError in my code. My code is:
> print(total + 5)
> Can you tell me what might be wrong?
>
> [LLM responds]
>
> Step 2: Thanks. Can you suggest a way to fix it?
>
> [LLM responds]
>
> Step 3: I tried that but it still gives an error. Any other suggestions?
>
> **LLM Output (expected):**
>
> Step 1: The error means 'total' is used before being defined.
> Step 2: Define 'total' first, e.g., total = 0
> Step 3: If you already tried that, make sure it's defined **before** the print statement:
> total = 0
> print(total + 5)



### 4. Chain-of-Thought for Debugging

**Description**: You guide the LLM through a logical sequence of steps to analyze the bug, encouraging it to reason step by step.

> **Prompt Example (to LLM):**
>
> Let's debug this step by step. I have the code:
> print(total + 5)
>
> Step 1: Is 'total' declared anywhere?
> Step 2: Is 'total' in the correct scope?
> Step 3: Is it being used before it's defined?
>
> Can you analyze each step and identify the issue?
>
> **LLM Output (expected):**
>
> Step 1: 'total' is not declared anywhere.
> Step 2: N/A, variable doesn't exist.  
> Step 3: It's being used before definition.
>
> Solution: Define 'total' before using it:
> total = 0
> print(total + 5)



### 5. Negative Prompting

**Description**: You tell the LLM explicitly what **not** to suggest, helping it avoid irrelevant or misleading solutions.

> **Prompt Example (to LLM):**
>
> I have this error:
> "NameError: name 'total' is not defined"
>
> Please help me fix it, but do **not** suggest:
> - Syntax errors
> - Missing semicolons
> - Problems with print() itself
>
> Focus only on issues with variable usage.
>
> **LLM Output (expected):**
>
> The error occurs because 'total' is being used before it is defined.  
> To fix it, define 'total' before the print statement:
> total = 0
> print(total + 5)

**Making Sense of Messy Code with an LLM**

**You’ve been given a block of messy, biology-related Python code. Your task is to use an LLM as a tool to understand and improve it.**

**Instructions:**

1. **Line-by-line annotation:** Ask the LLM to insert `# comments` explaining what each line of code does.
2. **High-level summary:** Write a clear description of the overall purpose of the script, including its inputs and outputs.
3. **Code critique:** Suggest concrete improvements such as clearer variable names, better error handling, and breaking the script into functions (modularization).
4. **Test run:** Use a sample `gene_expr.csv` file to run the code, then interpret the output (e.g., what is plotted, what genes get flagged, what files are produced).


````python
import pandas as pd, matplotlib.pyplot as plt, numpy as np, random

# script for bio expr
df = pd.read_csv("gene_expr.csv")
df = df.fillna(0)
genes = list(df.columns[1:])
samples = list(df['sampleID'])
Z = {}
for g in genes:
    vals = []
    for i in range(len(samples)):
        try:
            vals.append(float(df[g].iloc[i]))
        except:
            vals.append(0)
    Z[g]=vals

res = {}
for g in Z:
    m = np.mean(Z[g])
    s = np.std(Z[g])
    res[g] = (m, s, (m/s if s!=0 else 0))

sorted_genes = sorted(res.items(), key=lambda x:x[1][2], reverse=True)
top = sorted_genes[:10]

plt.figure(figsize=(9,5))
for tg in top:
    plt.plot(range(len(Z[tg[0]])), Z[tg[0]], label=tg[0])
plt.legend()
plt.title("top genes var/mean")
plt.savefig("out.png")

thresh = 1.5
flagged = []
for g in res:
    if res[g][2]>thresh and random.random()>0.2:
        flagged.append(g)

with open("flagged.txt","w") as f:
    for g in flagged:
        f.write(g+"\n")
````
